# Matplotlib

- 과학계산용 그래프 라이브러리
    - 선 그래프, 히스토그램, 산점도 등을 지원
- 그래프를 그리기 위해서 matplotlib의 pyplot을 이용

### 1. sin 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 준비
x = np.arange(0, 6, 0.1) # 0부터 6까지 0.1간격으로 배열 생성
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.show()
```

### 2. cos 그래프 그리기

```python
x = np.arange(0, 6, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x, sin_y, label="sin")
plt.plot(x, cos_y, linestyle="--", label="cos")

plt.title("sin, cos graph") # 제목

# 축 이름
plt.xlabel("x")
plt.ylabel("y")

plt.legend() # 범례 출력
plt.show()
```
